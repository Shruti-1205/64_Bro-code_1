```python
data['gender']=data['gender'].map({'Female':1, 'Male':2, 'Prefer_not_to_say':0})
data['age']=data['age'].map({'>45':4, '26-35':2, '36-45':3,'16-25':1,'Prefer_not_to_say':0})
data['geography']=data['geography'].map({'Suburban,Peri-urban':1, 'City,Metropolitan':2, 'Rural':0})
data['financial_situation']=data['financial_situation'].map({'food_regexpenses_nothingelse':3,
       'notenoughfood':1,
       'food_regexpenses_clothes_nothingelse':3,
       'food_nothingelse':2,
       'Prefer not to answer':0})
data['education']=data['education'].map({'Univ_college_degree_completed':4,
       'diploma_degreecompleted':5,
       'Secondary_highschoolcompleted':3,
       'Primary_school_completed':2, 'Postgraduate':6,
       'Prefer_not_to_say':0, 'No_formal_education':1})
data['employment_status']=data['employment_status'].map({
       'PartTime':3, 'FullTime':4, 'PartTime(Student)':2, 'Student':1,
       'Unemployed':0})
data['do_children_3_and_17_yrs_receive_regular_school_meals']=data['do_children_3_and_17_yrs_receive_regular_school_meals'].map({'1-2days':2, 'Yes':1, 'No':0, '3-4days':3})
data['are_there_teachers_at_scheduled_class_hours']=data['are_there_teachers_at_scheduled_class_hours'].map({'sometimes':1, 'Yes':2, 'No':0})
```

```python
import plotly.graph_objs as go
import plotly.offline as py
```

```python
def correlation_plot():

    correlation = data.corr()

    matrix_cols = correlation.columns.tolist()

    corr_array  = np.array(correlation)
    trace = go.Heatmap(z = corr_array,
                       x = matrix_cols,
                       y = matrix_cols,
                       colorscale='Viridis',
                       colorbar   = dict() ,
                      )
    layout = go.Layout(dict(title = 'Correlation Matrix for variables',
                            margin  = dict(r = 0 ,l = 100,
                                           t = 0,b = 100,
                                           ),
                            yaxis   = dict(tickfont = dict(size = 9)),
                            xaxis   = dict(tickfont = dict(size = 9)),
                           )
                      )
    fig = go.Figure(data = [trace],layout = layout)
    py.iplot(fig)
```

```python
correlation_plot()
```
